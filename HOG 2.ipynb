```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
import matplotlib.pyplot as plt
```

```python
img=imread('d1.jpg')
plt.axis("off")
plt.imshow(img)
print(img.shape)
```

```python
resized_img=resize(img,(128*10,64*10))
plt.axis("off")
plt.imshow(resized_img)
plt.show()
print(resized_img.shape)
```

```python
fd,hog_image=hog(img,orientations=10,pixels_per_cell=(2,2),cells_per_block=(4,4),visualize=True,channel_axis=2)
```

```python
#show image
print(fd.shape)
print(hog_image.shape)
plt.axis("off")
plt.imshow(hog_image,cmap='hot')
plt.show()
```
