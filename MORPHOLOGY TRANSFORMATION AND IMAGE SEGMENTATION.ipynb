#### MORPHOLOGY TRANSFORMATION AND IMAGE SEGMENTATION


```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

#### DILATION

```python
img=cv.imread('v2.jpg',0)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,225,cv.THRESH_BINARY)
kernel=np.ones((3,3),np.uint8)
dilation=cv.dilate(mask,kernel,iterations=1)
erode=cv.erode(mask,kernel,iterations=1)
titles=['image','mask','dilate','erode']
images=[img,mask,dilation,erode]
for i in range(len(images)):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
